```python
for col in ['temp', 'atemp', 'hum','windspeed','casual','registered','cnt']:
    q1= final_dataset [col].quantile(0.25)
    q3= final_dataset [col].quantile(0.75)
    iqr =q3-q1 
    lower_bound =q1-1.5 *iqr
    upper_bound = q3 + 1.5 *iqr 
    
```

```python
final_dataset[col]= np.clip(final_dataset[col],lower_bound,upper_bound)
```

```python
final_dataset[col]
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x=['final_dataset[col]','values'])
plt.title('Violin Plot to Detect Outliers')
plt.show()
```
